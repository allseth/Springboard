```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('healthcare-dataset-stroke-data.csv')
data.head()
```

```python
drop_data = data.dropna()
drop_data.head()
```

```python
stroke_data = drop_data.drop(columns=['id'])
stroke_data.head()
```

```python
# Create a pie chart to visualize stroke percentage
colors = ('red', 'gray')
explode = (0.05,0.05)
wedge = {'edgecolor':'black'}
stroke_percentage = stroke_data['stroke'].value_counts()
yes_no = ['Stroke', 'No Stroke']
stroke_percentage.plot(kind='pie', autopct='%0.1f%%', labels= yes_no, colors=colors, explode=explode, wedgeprops=wedge)
plt.title('Stroke Percentage', loc='center', color='black', fontsize='15', fontweight='bold')
plt.ylabel('')
plt.show()
```

```python
sns.kdeplot(data=stroke_data, x='age',hue='stroke', palette='magma', fill=True, common_norm=False, alpha=.2, linewidth=0.5)
plt.title('Correlation Between Stroke and Age', fontsize='15', fontweight='bold')
plt.xlabel('Age')
plt.show()
```

/Based on the plot above, the chance of experiencing a stroke increases at the age of 40.

```python
sns.kdeplot(data=stroke_data, x='avg_glucose_level', hue='stroke', palette='pastel', fill=True, common_norm=False, alpha=0.5, linewidth=1)
plt.title('Correlation Between Glucose Level and Stoke', fontsize='15', fontweight='bold')
plt.xlabel('Glucose Level')
plt.show()
```

```python
sns.kdeplot(data=stroke_data, x='bmi', hue='stroke', palette='husl', fill=True, common_norm=False, alpha=0.5, linewidth=1)
plt.title('Correlation Between Body Mass Index and Stroke', fontsize='15', fontweight='bold')
plt.xlabel('BMI')
plt.show()
```

```python
sns.catplot(data=stroke_data, x='gender', hue='stroke', kind='count', palette='Set2')
plt.title('Distribution of Stroke by Gender',fontsize='15', fontweight='bold')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
sns.catplot(data=stroke_data, x='work_type', hue='stroke', kind='count', palette='Set2')
plt.title('Distribution of Stroke by Work Type',fontsize='15', fontweight='bold')
plt.xlabel('Work Type')
plt.ylabel('Count')
plt.show()
```

```python
sns.catplot(data=stroke_data, x='ever_married', hue='stroke', kind='count', palette='Set2')
plt.title('Distribution of Stroke in Ever Married',fontsize='15', fontweight='bold')
plt.xlabel('Ever Married')
plt.ylabel('Count')
plt.show()
```

```python
sns.catplot(data=stroke_data, x='hypertension', hue='stroke', kind='count', palette='Set2')
plt.title('Distribution of Stroke by Hypertension',fontsize='15', fontweight='bold')
plt.xlabel('Hypertension')
plt.ylabel('Count')
plt.show()
```

```python
sns.catplot(data=stroke_data, x='heart_disease', hue='stroke', kind='count', palette='Set2')
plt.title('Distribution of Stroke by Heart Disease',fontsize='15', fontweight='bold')
plt.xlabel('Heart Disease')
plt.ylabel('Count')
plt.show()
```

```python
sns.catplot(data=stroke_data, x='Residence_type', hue='stroke', kind='count', palette='Set2')
plt.title('Distribution of Stroke by Residence Type',fontsize='15', fontweight='bold')
plt.xlabel('Residence Type')
plt.ylabel('Count')
plt.show()
```

```python
sns.catplot(data=stroke_data, x='smoking_status', hue='stroke', kind='count', palette='Set2')
plt.title('Distribution of Stroke by Smoking Status',fontsize='15', fontweight='bold')
plt.xlabel('Smoking Status')
plt.xticks(rotation=45)
plt.ylabel('Count')
plt.show()
```

```python
sns.pairplot(data=stroke_data, hue='stroke')
plt.show()
```

```python
heatmap = sns.heatmap(stroke_data[['age', 'avg_glucose_level', 'bmi']].corr(), vmin=-1, vmax=1, annot=True, cmap='Blues')
plt.title('Correlation Between Age, Average Glucose Level, and BMI',fontsize='15', fontweight='bold')
```

```python
Conclusions

-Age and BMI have strong correlations
-Residence type did not have a strong correlation with stroke
-Gender did not have a strong correlation with stroke
-Weak correlation between stroke and smoking status
```
