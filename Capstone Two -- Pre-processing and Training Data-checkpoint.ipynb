### R-squared (Coefficient of Determination)
Measuring the proportion of variance in the dependent variable.

```python
#Calculate the R-squared
def r_squared(y, ypred):
    ybar = np.sum(y) / len(y) #yes, we could use np.mean(y)
    sum_sq_tot = np.sum((y - ybar)**2) #total sum of squares error
    sum_sq_res = np.sum((y - ypred)**2) #residual sum of squares error
    R2 = 1.0 - sum_sq_res / sum_sq_tot
    return R2
```

```python
#Make predictions by creating an array of length training set
y_tr_pred_ = train_mean * np.ones(len(y_train))
y_tr_pred_[:5]
```

```python
#Make predictions by creating an array of length test set
y_te_pred = train_mean * np.ones(len(y_test))
r_squared(y_test, y_te_pred)
```

### Mean Absolute Error

```python
#Calculate the Mean Absolute Error
def mae(y, ypred):
    abs_error = np.abs(y - ypred)
    mae = np.mean(abs_error)
    return mae

print(f'The mean absolute error for the training set is {mae(y_train, y_tr_pred)}')
print(f'The mean absolute error for the testing set is {mae(y_test, y_te_pred)}')
```

### Mean Squared Error

```python
#Calculate the Mean Squared Error
def mse(y, ypred):
    sq_error = (y - ypred)**2
    mse = np.mean(sq_error)
    return mse

print(f'The mean squared error for the training set is {mse(y_train, y_tr_pred)}')
print(f'The mean squared error for the testing set is {mse(y_test, y_te_pred)}')
```

```python
#Define the pipeline
RF_pipe = make_pipeline(
    SimpleImputer(strategy='median'),
    StandardScaler(),
    RandomForestRegressor(random_state=1255)
)
RF_pipe.fit(X_train, y_train)
```

```python
rf_default_cv_results = cross_validate(RF_pipe, X_train, y_train, cv=5)
rf_cv_scores = rf_default_cv_results['test_score']
rf_cv_scores
```
